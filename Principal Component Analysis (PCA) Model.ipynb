```python
#Problem Statement:
"""Suppose we have a dataset of student scores on three 
different subjects: Math, Science, and English. We want to apply 
Principal Component Analysis (PCA) to reduce the dimensionality 
of this dataset while retaining the most important information. 
Our goal is to represent the students' performance in a lower-dimensional 
space using the first principal component."""
```

```python
# step 1 - importing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
```

```python
# step no 2 - data define
data = {'Math Score': [80,75,90,85,70,78,92,88,72,81], 'Science Score':[85,78,92,88,75,80,95,90,76,84], 'English Score':[90,80,88,91,82,85,87,89,80,87]}
```

```python
# step 3 - creating a DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
# step no 4 - Performing PCA and fitting
pca = PCA(n_components=1)
principal_component=pca.fit_transform(df)
```

```python
# step 5 - principal componebt scores to dataframe
df['Principal_Component']=principal_component
```

```python
print(df['Principal_Component'])
print(df)
```

```python
# step no 6 - visualization
# Math Score  Science Score  English Score
math_Score = df['Math Score']
science_Score = df['Science Score']
english_Score = df['English Score']
pc_scores = df['Principal_Component']
fig=plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(math_Score, science_Score,english_Score)
ax1.set_xlabel('Math Score')
ax1.set_ylabel('Science Score')
ax1.set_zlabel('English Score')
ax1.set_title('Original Score')
# adding subplot

ax2 = fig.add_subplot(122)
ax2.scatter(pc_scores, np.zeros_like(pc_scores))
ax2.set_xlabel('Principal Component')
ax2.set_title('First Principal Component')
plt.tight_layout()
plt.show()
```
